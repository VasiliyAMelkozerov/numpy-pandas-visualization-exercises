## 2 Is there a correlation between petal length and petal width? Use http://guessthecorrelation.com/ as a hint to how we could visually determine if a correlation exists between two numeric columns.

```python
sns.relplot(data = iris, 
            x="petal_length", 
            y="petal_width",
            hue ="species")
```

```python
sns.pairplot(data = iris, corner=True)
```

```python
sns.heatmap(iris.corr(), 
            annot=True, 
#             fmt='0.01f',
           cmap='Blues')
```

```python
sns.heatmap(iris.corr(), 
            annot=True, 
            fmt="0.3f",
           cmap='Blues')
```

### we can say that petal width and sepal length have a strong correlation to petal length

## 3 Would it be reasonable to predict species based on sepal width and sepal length? For this, you'll visualize two numeric columns through the lense of a categorical column.

```python
sns.relplot(data = iris, 
            x = "sepal_length", 
            y = "sepal_width",
           hue="species")
```

```python
sns.catplot(data=iris)
sns.catplot(data = iris, x = 'sepal_length', y ='sepal_width', hue='species')
# sns.catplot(data = iris, x ='sepal_width', y = 'sepal_length')
```

```python
sns.pairplot(data = iris, hue="species")
```

### yes it would be reasonable to predict species based on sepal width and sepal length
